### Exercise 1: Data Cleaning with VS Code

This exercise is accompanied by the *AAPL.csv* file containing historical stock prices for Apple Inc. Your task is to clean the dataset in VS Code so that the following Python code correctly produces a visualization of the closing prices over time.
Do **not** modify or write any code for this exercise. Focus solely on cleaning the CSV file.
Here are a few recommended steps for this exercise:

1. Install *pandas* and *matplotlib* using pip.
2. Add the *Rainbow CSV* extension to VS Code for better CSV handling.
3. Use the keyboard shortcuts from the lecture to format the CSV file properly until the provided code runs without errors. *Hint*: A common date format in coding is `YYYY-MM-DD`, e.g. `2023-06-15`.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("AAPL.csv", sep=",")
df["Close"] = df["Close"].astype(float)
df["Date"] = pd.to_datetime(df["Date"])
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Close"], label="Close Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("Apple Closing Stock Price Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````
